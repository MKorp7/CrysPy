```python
PATH_TO_NSP3_RESULT = "NetSurfP-3.0_standalone/nsp_result_9_1_c/01/01.csv"  # created by merging of all csv results from NetSurfP-3.0
PATH_TO_CLASSIC_FEATURES = "9_1.csv"# created with notebook "1_standard_features_calculation.ipynb"
WHERE_TO_SAVE_OUTPUT = "9_1_full.csv" #There the result will be saved
```

```python
import pandas as pd
```

```python
df_netsurf = pd.read_csv(PATH_TO_NSP3_RESULT)
df_netsurf['id'] = df_netsurf['id'].str.replace('>', '', regex=False)
```

```python
df_features = pd.read_csv(PATH_TO_CLASSIC_FEATURES) 
```

```python
df_features
```

```python
df_netsurf['id'] = df_netsurf['id'].str.split('|').str[1]
```

```python
from tqdm import tqdm
results = []

for protein_id in tqdm(df_netsurf['id'].unique()):
    df = df_netsurf[df_netsurf['id'] == protein_id]
    # Kyte-Doolittle scale
    hydropathy_index = {
        'A': 1.8, 'R': -4.5, 'N': -3.5, 'D': -3.5, 'C': 2.5, 
        'Q': -3.5, 'E': -3.5, 'G': -0.4, 'H': -3.2, 'I': 4.5, 
        'L': 3.8, 'K': -3.9, 'M': 1.9, 'F': 2.8, 'P': -1.6, 
        'S': -0.8, 'T': -0.7, 'W': -0.9, 'Y': -1.3, 'V': 4.2
    }
    df['hydropathy'] = df[' seq'].map(hydropathy_index)

    M = df_features[df_features['ID'] == protein_id]['Molecular_weight'].values[0]
    sum_asa = df[' asa'].sum()
    SR = sum_asa / (6.3 * (M ** 0.73))
    
    weigth_hydropathy = (df['hydropathy'] * df[' rsa']).sum()
    
    sum_rsa = df[' rsa'].sum()
    gravy_index_weighted = weigth_hydropathy / sum_rsa if sum_rsa > 0 else 0
    weigth_disorder = (df[' disorder'] * df[' rsa']).sum() / sum_rsa if sum_rsa >0 else 0
    weighted_aa_composition_full = df.groupby(' seq')[' rsa'].sum().reset_index()
    weighted_aa_composition_full.columns = ['amino_acid', 'weighted_count']
    
    weighted_aa_composition_dict = weighted_aa_composition_full.set_index('amino_acid')['weighted_count'].to_dict()
    weighted_aa_composition_formatted = {f'count of surface {aa}': count/sum_rsa for aa, count in weighted_aa_composition_dict.items()}
    result = {
        "Protein_ID": protein_id,
        "ASA": sum_asa,
        "ASA_avg" : sum_asa/ len(df[' asa']),
        "aa_on_surface": len(df[df[' rsa'] >= 0.25]),
        "RSA": sum_rsa,
        "Weighted_disorder": weigth_disorder,
        "RSA_avg": sum_rsa/ len(df[' rsa']),
        "disorder_inside": len(df[(df[' rsa'] < 0.25) & (df[' disorder'] > 0.5)]),
        "Surface_ruggedness": SR,
        "Weighted_GRAVY": gravy_index_weighted,
    }
    result.update(weighted_aa_composition_formatted)
    results.append(result)

results
```

```python
df_merged = pd.merge(df_features, pd.DataFrame(results), left_on='ID', right_on='Protein_ID', how='left')
```

```python
df_merged.to_csv(WHERE_TO_SAVE_OUTPUT, index=False)
```
